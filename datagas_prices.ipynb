```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('gas_prices.csv')
gas 
```

```python
plt.title('Price of fuel in US$')

list_country = ['Japan', 'USA', 'France', 'Mexico', 'Germany']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label=[country], marker='.')


plt.xticks(gas['Year'][::2])#de dois em dois anos pra me mostrar. 

plt.xlabel('Years') 
plt.ylabel('US$')

plt.legend()
plt.savefig('country.pdf')
plt.show()
```
